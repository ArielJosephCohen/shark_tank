```python
from autogluon import TabularPrediction as task
import pprint
from IPython.display import display
import pandas as pd
```

```python
df=pd.read_csv('shark_tank.csv')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split((df.drop('deal',axis=1)),df.deal)

train = pd.concat([X_train,y_train],axis=1)

test = pd.concat([X_test,y_test],axis=1)
```

```python
label_column = 'deal'
print("Possible values of target variable (with occurrence counts): \n", train[label_column].value_counts().to_string())
```

```python
predictor = task.fit(train_data=train, label=label_column)
```

```python
y_test = test[label_column]
test_data_nolab = test.drop(labels=[label_column],axis=1)
display(test_data_nolab.head())
```

```python
y_pred = predictor.predict(test_data_nolab)
perf = predictor.evaluate_predictions(y_true=y_test, y_pred=y_pred, auxiliary_metrics=True)
print("Predictions:\n", y_pred)
```
